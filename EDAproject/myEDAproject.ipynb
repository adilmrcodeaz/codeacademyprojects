```python
import seaborn as sns
import matplotlib.pyplot as plt

# Veri setini yÃ¼kleme
df = pd.read_csv('myEDAproject.csv')

# Subplots oluÅŸturma
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# Ä°lk Ã§izgi grafiÄŸi
sns.lineplot(data=df.head(50), x='productCode', y='orderDate', ax=ax1)

# Ä°kinci Ã§izgi grafiÄŸi
sns.lineplot(data=df.head(50), x='productCode', y='orderLineNumber', ax=ax2)

# x eksenindeki etiketleri dÃ¶ndÃ¼rme
plt.xticks(rotation=180)

# Grafikleri gÃ¶sterme
plt.show()
```
